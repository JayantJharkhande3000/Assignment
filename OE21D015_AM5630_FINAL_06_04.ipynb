```python
print(   "\t ########################################################################### \t"  )
print(   "\t #############   Elliptical Equation 2D LINE Gauss seidal solution ######## \t"  )
print(   "\t ########################################################################### \t"  )


# Defining the A (n+1 time level),B(n+1 time level) and C(boundary condition) matrix 
A=np.zeros((nx-2,nx-2))
B=np.zeros((nx-2,nx-2))
C=np.zeros((nx-2,nx-2))
P=np.zeros((nx-2,nx-2))
D=np.zeros((nx-2,1))
E=np.zeros((nx-2,1))
F=np.zeros((nx-2,1))
G=np.zeros((nx-2,1))
H=np.zeros((nx-2,1))
I=np.zeros((nx-2,1))
M=np.zeros((ny,nx))
N=np.zeros((ny,nx))
O=np.zeros((ny,nx))

# Array for calculation the temperature variation at specific distance
#  x represent the distance and y1 to y5 represent the specific point in the model
x= np.zeros(ny)
y1= np.zeros(ny)
y2= np.zeros(ny)
y3= np.zeros(ny)
y4= np.zeros(ny)
y5= np.zeros(ny)

# defining the error at any value
start=time.time()
Error=10
k=0

# Loop calculation Part
for i in range (0,nx-2):
    for j in range (0,nx-2):
        if i==j:
            A[i][j]=-4
            B[i][j]=-1
            C[i][j]=-1
        if j==i+1:
            A[i][j]=1
        if j==i-1:
            A[i][j]=1
Error=10
while Error>.01:
    Error=0
    for j in range(1, ny-1):
        for i in range(1, nx-1):
            E[i - 1][0] = O[j + 1][i]     #E is previous time step matrix
            F[i - 1][0] = N[j - 1][i]     #F is present time step matrix
        G = np.dot(B, E)
        H = np.dot(C, F)
        I = np.add(G, H)
        D = np.linalg.solve(A, I)
        for i in range (1,30):
            N[j][i] = D[i - 1][0]
            Error=Error+abs(N[j][i]-O[j][i])
    for i in range(0, 31):
        for j in range(0, 41):
            if j == 0:
                M[j][i] = 40
                N[j][i] = 40
                O[j][i] = 40
            if j == 40:
                M[j][i] = 10
                N[j][i] = 10
                O[j][i] = 10
    k=k+1
    if k==1:
        Error=10
    for j in range(1, 40):
        for i in range(1, 30):
            O[j][i] = N[j][i]
print("\n \t  the total number of iteration is :", k)

end=time.time()
print(f"\n \t Runtime of the program is {end - start}")
print("\n\n")


# Visualization Part 
for i in range (0,ny):
    x[i]=i*.01
    y1[i] = O[i][5]
    y2[i] = O[i][10]
    y3[i] = O[i][15]
    y4[i] = O[i][20]
    y5[i] = O[i][25]

plt.plot(x, y1, color='maroon', label='.05')
plt.plot(x, y2, color='crimson', label='.10')
plt.plot(x, y3, color='hotpink', label='.15')
plt.plot(x, y4, color='pink', label='.20')
plt.plot(x, y5, color='tomato', label='.25')
plt.title(" For all y nodes at dx=0.5 at steady state ")
plt.xlabel(" Plate Distance (m) ")
plt.ylabel(" Temperature distribution ")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

import time
 
# we initialize the variable start
# to store the starting time of
# execution of program
start = time.time()


print(   "\t ########################################################################### \t"  )
print(   "\t ############### Elliptical Equation 2D Point Successive over relaxation method ###################### \t"  )
print(   "\t ########################################################################### \t"  )
    

L = 0.3
print("\n\tThe lenght is ", str(L))
n = 30
print("\n\t x direction numeber of nodes are ", str(n))

H = 0.4
print("\n\t The height is ", str(H))
m = 40
print("\n\t y direction numeber of nodes are ", str(m))


Tinitial = 0
print("\n\tThe intial temperature of the system is ", str(Tinitial))
Tl = 0
print("\n\tThe temperature at the left boundary of the system is ", str(Tl))
Tr = 0
print("\n\tThe temperature at the right boundary of the system is ", str(Tr))
Tb = 40
print("\n\tThe temperature at the bottom boundary of the system is ", str(Tb))
Tt = 10
print("\n\tThe temperature at the top boundary of the system is ", str(Tt))

dx = L/n
print("\n\t The grid lenght dx is  : ", str(dx))

dy = H/m
print("\n\t The grid lenght dy is  : ", str(dy))


thermal_diffusivity = 11.234*10**(-5)
print("\n\t thermal diffusivity : ", str(thermal_diffusivity))

thermal_conductivity = 380
print("\n\t thermal Conductivity : ", str(thermal_conductivity))

colorinterpolation = 50 
colourMap = plt.cm.jet


MaxIter =  300
print("\n\t MaxIteration : ", str(MaxIter))

lenX = n
lenY = m


beta = float(1)
relaxation_coff = float(1.5)
gamma = float(relaxation_coff/(2*(1+beta**2))) 
print("\n\t gamma : ", str(gamma))



delta = dx

colorinterpolation = 50 
colourMap = plt.cm.jet
T = np.empty((lenX, lenY))
T.fill(Tinitial)

print("the shape of T matrix",T.shape)

print("\n ####################################3333 \n")

T[(lenX-1):,:]=Tt
T[:1, :] = Tb
T[:,(lenY-1):]=Tr
T[:,:1]=Tl

X, Y = np.meshgrid(np.arange(0,lenY),np.arange(0,lenX))
plt.figure(1)
plt.figure(figsize=(8,8))
plt.title("Temperature  distribution Elliptical Equation 2D Point Gauss seidal solution")
plt.xlabel('Grid node in x direction')
plt.ylabel('Grid node in y direction')

plt.contourf(X,Y, T, colorinterpolation, cmap = colourMap)

plt.colorbar()
plt.show()


print(T.shape)

print("\n ####################################3333 \n")


print("\n solution \n")
for iterarionn in range(0, MaxIter):
  for i in range(1,lenX-1):
    for j in range(1, lenY - 1):
      T[i,j] = (1-relaxation_coff)*T[i,j] +  gamma*(T[i+1][j] + T[i-1][j] + (beta**2)*(T[i][j+1] + T[i][j-1]))

print("\n Iteration is finished\n")     

X, Y = np.meshgrid(np.arange(0,lenY),np.arange(0,lenX))
plt.figure(1)
plt.figure(figsize=(8,8))
plt.title("Temperature  distribution Elliptical Equation 2D Point Successive over relaxation solution at 300 iteration")
plt.xlabel('Grid node in x direction')
plt.ylabel('Grid node in y direction')
plt.contourf(X,Y, T, colorinterpolation, cmap = colourMap)

plt.colorbar()
plt.show()


end = time.time()
 
# difference of start and end variables
# gives the time of execution of the
# program in between
print("\n\t The time of execution of above program is (sec):", end-start)
```
